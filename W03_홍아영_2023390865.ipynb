```python
import seaborn as sns
import matplotlib.pyplot as plt


attrition_by_job_satisfaction = df.groupby('JobSatisfaction')['Attrition_numeric'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=attrition_by_job_satisfaction, x='JobSatisfaction', y='Attrition_numeric')
plt.title('Attrition Rate by Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Attrition Rate')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Low', 'Medium', 'High', 'Very High'])
plt.show()
```
